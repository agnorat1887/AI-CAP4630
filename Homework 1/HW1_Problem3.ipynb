#Import and reshape images

```
from keras.datasets import mnist

(train_images_original, train_labels_original), (test_images_original, test_labels_original) = mnist.load_data()

import matplotlib.pyplot as plt
import numpy as np

train_images = train_images_original.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = test_images_original.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

#Sort the same numbers into same array

```
train_0 = [] 
train_1 = []
train_2 = []
train_3 = []
train_4 = []
train_5 = []
train_6 = []
train_7 = []
train_8 = []
train_9 = []

for i in range(60000):
  if(train_labels_original[i] == 0):
    train_0.append(train_images[i])
  elif(train_labels_original[i] == 1):
    train_1.append(train_images[i])
  elif(train_labels_original[i] == 2):
    train_2.append(train_images[i])
  elif(train_labels_original[i] == 3):
    train_3.append(train_images[i])
  elif(train_labels_original[i] == 4):
    train_4.append(train_images[i])
  elif(train_labels_original[i] == 5):
    train_5.append(train_images[i])
  elif(train_labels_original[i] == 6):
    train_6.append(train_images[i])
  elif(train_labels_original[i] == 7):
    train_7.append(train_images[i])
  elif(train_labels_original[i] == 8):
    train_8.append(train_images[i])
  elif(train_labels_original[i] == 9):
    train_9.append(train_images[i])
```

#Find average of all numbers

```
train_0 = np.array(train_0)
av_0 = np.average(train_0, 0)

train_1 = np.array(train_1)
av_1 = np.average(train_1, 0)

train_2 = np.array(train_2)
av_2 = np.average(train_2, 0)

train_3 = np.array(train_3)
av_3 = np.average(train_3, 0)

train_4 = np.array(train_4)
av_4 = np.average(train_4, 0)

train_5 = np.array(train_5)
av_5 = np.average(train_5, 0)

train_6 = np.array(train_6)
av_6 = np.average(train_6, 0)

train_7 = np.array(train_7)
av_7 = np.average(train_7, 0)

train_8 = np.array(train_8)
av_8 = np.average(train_8, 0)

train_9 = np.array(train_9)
av_9 = np.average(train_9, 0)


```

#Reshape into correct image dimensions

```
av_0 = av_0.reshape(28, 28)
av_1 = av_1.reshape(28, 28)
av_2 = av_2.reshape(28, 28)
av_3 = av_3.reshape(28, 28)
av_4 = av_4.reshape(28, 28)
av_5 = av_5.reshape(28, 28)
av_6 = av_6.reshape(28, 28)
av_7 = av_7.reshape(28, 28)
av_8 = av_8.reshape(28, 28)
av_9 = av_9.reshape(28, 28)

av = []

for i in range(10):
  av.append(av_0)
  av.append(av_1)
  av.append(av_2)
  av.append(av_3)
  av.append(av_4)
  av.append(av_5)
  av.append(av_6)
  av.append(av_7)
  av.append(av_8)
  av.append(av_9)
```

#Show average of numbers

```
fig = plt.figure(figsize=(8, 20))
rows = 5
columns = 2
for i in range(1, columns*rows+1):
  fig.add_subplot(rows, columns, i)
  plt.imshow(av[i-1])
plt.show()
```
