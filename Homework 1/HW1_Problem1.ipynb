#Local Minimum

```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

A function with a local minimum

```
def f1(x, y):
  return x**2 + y**2
```

Graph for the local minimum function

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

fig = plt.figure()
ax = fig.add_subplot(1, 2, 1, projection = '3d')
ax.contour3D(X, Y, Z, 30)
ax.scatter3D(0, 0, 0, s=500)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

axe = fig.add_subplot(1, 2, 2)
axe.contour(X, Y, Z, colors='black')
axe.scatter(0, 0)

plt.show()
```

#Local Maximum

```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

A function with a Local Maximum

```
def f2(x, y):
  return -x**2 - y**2
```

A graph for the Local Maximum

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)

fig = plt.figure()
ax = fig.add_subplot(1, 2, 1, projection = '3d')
ax.contour3D(X, Y, Z, 30)
ax.scatter3D(0, 0, 0, s=500)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

axe = fig.add_subplot(1, 2, 2)
axe.contour(X, Y, Z, colors='black')
axe.scatter(0, 0)

plt.show()
```

#Saddle Point

```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

A function with a saddle point

```
def f3(x, y):
  return x**2 - y**2
```

A graph with a saddle point

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)

fig = plt.figure()
ax = fig.add_subplot(1, 2, 1, projection = '3d')
ax.contour3D(X, Y, Z, 30)
ax.scatter3D(0, 0, 0, s=500)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

axe = fig.add_subplot(1, 2, 2)
axe.contour(X, Y, Z, colors='black')
axe.scatter(0, 0)

plt.show()
```
